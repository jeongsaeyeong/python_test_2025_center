```python
# 중첩 루프
breads = ['호밀빵', "위트", "화이트"]
meats = ["미트볼", "소시지", "닭가슴살"]
vegis = ["양상추", "토마토", "오이"]
sauces = ["마요네즈", "허니 머스타드", "칠리"]

print("우리가게 조합 가능한 샌드위치 종류")
for b in breads:
    for m in meats:
        for v in vegis:
            for s in sauces:
                print(b + " +  " + m + " + " + v + " + " + s)
```

```python
# 중첩 루프 - 개선 코드 1 (참고)
# list comprehension 활용
# 가독성 개선
breads = ['호밀빵', "위트", "화이트"]
meats = ["미트볼", "소시지", "닭가슴살"]
vegis = ["양상추", "토마토", "오이"]
sauces = ["마요네즈", "허니 머스타드", "칠리"]

print("우리가게 조합 가능한 샌드위치 종류")

# 리스트를 하나로 합쳐서 처리
combinations = [(b, m, v, s) for b in breads for m in meats for v in vegis for s in sauces]

# 모든 조합 출력
for combo in combinations:
    print(" + ".join(combo))
```

```python
# 중첨 루프 - 개선 코드 2 (참고)
# itertools.product 활용
# 내장함수
from itertools import product

breads = ['호밀빵', "위트", "화이트"]
meats = ["미트볼", "소시지", "닭가슴살"]
vegis = ["양상추", "토마토", "오이"]
sauces = ["마요네즈", "허니 머스타드", "칠리"]

print("우리가게 조합 가능한 샌드위치 종류")

# itertools.product를 사용하여 모든 조합 생성
for b, m, v, s in product(breads, meats, vegis, sauces):
    print(f"{b} + {m} + {v} + {s}")
```

```python
# 중첩 루프 - 개선 코드 3 (참고)
# 재귀적 함수 호출을 통해 데이터 구조를 미리 결합
# 단점? => 결합된 데이터를 저장할 저장 공간 증가
breads = ['호밀빵', "위트", "화이트"]
meats = ["미트볼", "소시지", "닭가슴살"]
vegis = ["양상추", "토마토", "오이"]
sauces = ["마요네즈", "허니 머스타드", "칠리"]

# 함수로 각 재료 조합 생성
def generate_combinations(lists):
    if not lists:
        return [[]]
    result = []
    first_list = lists[0]
    remaining_lists = lists[1:]
    
    for item in first_list:
        for combo in generate_combinations(remaining_lists):
            result.append([item] + combo)
    
    return result

# 리스트들 순서를 전달
combinations = generate_combinations([breads, meats, vegis, sauces])
print(combinations)

# 결과 출력
print("우리가게 조합 가능한 샌드위치 종류")
for combo in combinations:
    print(" + ".join(combo))
```

```python
# 루프 제어문 - break
coffee = 10
money = 30
while money:
    print("돈이 지불되어 커피를 판매합니다.")
    money -= 2
    coffee -= 1
    print("남은 커피의 수량은 %d입니다" % coffee)
    if not coffee:
        print("커피가 소진되어 판매를 중단합니다.")
        break
print("남은 돈은 %d입니다" % money)
```

```python
# 루프 제어문 - break, continue 빈번한 사용
def process_numbers(numbers):
    for num in numbers:
        if num < 0:
            continue  # 음수는 무시
        if num % 2 == 0:
            print(f"{num}은(는) 짝수입니다.")
            if num > 100:
                print(f"{num}은(는) 100보다 큽니다.")
                break  # 100보다 큰 짝수가 있으면 루프 종료
        else:
            if num % 3 == 0:
                print(f"{num}은(는) 3의 배수입니다.")
                continue  # 3의 배수는 나머지 처리를 건너뜀
            if num > 50:
                print(f"{num}은(는) 50보다 큽니다.")
            else:
                print(f"{num}은(는) 작은 홀수입니다.")

process_numbers([1, 20, 50, 100, 230])
```
