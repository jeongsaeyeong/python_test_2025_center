### 11. Regular Expression

```python
"""
텍스트 - 정규식 1 - search( ) 
정규식 - 규칙을 이용한 문자 추출
"""
import re

txt1 = "Life is too short, you need python."
txt2 = "The best moments of my life."

# 문장 안에 Life가 있는가 검사함
print(re.search('Life', txt1)) 
print(re.search('Life', txt2))
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출
"""
match = re.search('Life', txt1)
print(match.group())
print(match.start())
print(match.end())
print(match.span())
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출
"""
match = re.search('Life', txt2)
print(match.group())
print(match.start())
print(match.end())
print(match.span())
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출: 메타 문자 활용 1
"""
# 문장 안에 Life 또는 life가 있는가 검사
print(re.search('Life|life', txt2))    
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출: 메타 문자 활용 2
"""
# 문장 안에 Life 또는 life가 있는가 검사함
print(re.search('[Ll]ife', txt2)) 
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출: 메타 문자 활용 5
"""

import re

print(re.search('A..A', 'ABA'))      # 조건에 부합하지 않음
print(re.search('A..A', 'ABBA'))     # 조건에 부합
print(re.search('A..A', 'ABBBA'))    # 조건에 부합하지 않음
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출: 메타 문자 활용 6
"""

import re

# A: 문자열에서 'A' 체크
# B*: 0개 이상의 'B' 체크
# 즉, B는 없어도 매치됨
print(re.search('AB*', 'A'))         
print(re.search('AB*', 'AA') )       
print(re.search('AB*', 'J-HOP'))
print(re.search('AB*', 'X-MAN'))
print(re.search('AB*', 'CABBA'))
print(re.search('AB*', 'CABBBBBA'))
print(re.search('AB*', 'CBBBBBA'))
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출: 메타 문자 활용 7
"""

import re

# A: 문자 'A'를 체크
# B?: 0개 또는 1개의 'B'를 체크
# 즉, 'B'가 없거나 1개만 있어야 매치됨
print(re.search('AB?', 'A'))     # A    
print(re.search('AB?', 'AA'))    # A   
print(re.search('AB?', 'J-HOP')) # None
print(re.search('AB?', 'X-MAN')) # A
print(re.search('AB?', 'CABBA')) # AB
print(re.search('AB?', 'CABBBBBA')) # AB
print(re.search('AB?', 'CBBBBBA'))  # A
print(re.search('AB?', 'AAABBBB'))  # A
print(re.search('AB?', 'ABBBABBBB')) # AB
```

```python
""" 
정규식 - 규칙을 이용한 문자 추출: 메타 문자 활용 8
"""

import re

# A: 문자 'A'를 체크
# B+: 1개 이상의 'B'를 체크
# 즉, 'B'는 반드시 최소 1개 있어야 함
print(re.search('AB+', 'A'))         # None
print(re.search('AB+', 'AA'))        # None
print(re.search('AB+', 'J-HOP'))     # None
print(re.search('AB+', 'X-MAN'))     # None
print(re.search('AB+', 'CABBA'))     # ABB
print(re.search('AB+', 'CABBBBBA'))  # ABBBBB
print(re.search('AB+', 'CBBBBBA'))   # None
print(re.search('AB+', 'AAABBBB'))   # ABBBB
print(re.search('AB+', 'ABBBABBBB')) # ABBB
```

```python
""" 
정규식 - 함수 활용
"""
import re

# 테스트할 문자열
test_string = "abcde 43535 fg rj3_343 hi 237 jk7mn 3opqr *+=$#@ st uvwxy 1 zzzzz ssssss"

# 단어 문자 매칭 (단어 문자: 알파벳 대소문자, 숫자, 언더스코어 포함 / 공백 제외)
pattern_1 = r'\w{5}'

# 단어 문자가 아닌 문자에 매칭 (특수문자 및 공백)
pattern_2 = r'\W{5}'

# 공백 문자(스페이스, 탭, 줄바꿈 등)에 매칭
# \b\s 의미: 단어가 끝(경계)나고, 공백이 이어지는 패턴 => 즉, 단어의 끝 경계에 있는 공백 매칭
# \s\b 의미: 공백 이후 단어가 시작(경계)되는 패턴 => 즉, 단어의 시작 경계에 있는 공백 매칭 
pattern_3 = r'\b\s\b'

# 문자 매칭 (공백 제외)
pattern_4 = r'\b\S{5}\b'

# (하나 이상인) 숫자 매칭
pattern_5 = r'\b\d[0-9]+\b'

# 숫자가 아닌 문자 매칭 (공백 포함)
pattern_6 = r'\b\D{3}\b'

# 정규식으로 다섯 글자 패턴을 찾음
matches_1 = re.findall(pattern_1, test_string)
matches_2 = re.findall(pattern_2, test_string)
matches_3 = re.findall(pattern_3, test_string)
matches_4 = re.findall(pattern_4, test_string)
matches_5 = re.findall(pattern_5, test_string)
matches_6 = re.findall(pattern_6, test_string)

# 결과 출력
print("Matched patterns 1:", matches_1)
print("Matched patterns 2:", matches_2)
print("Matched patterns 3:", matches_3)
print("Matched patterns 4:", matches_4)
print("Matched patterns 5:", matches_5)
print("Matched patterns 6:", matches_6)
```

```python
""" 
정규식 - 함수 활용
"""

# 1) re 함수에서 직접 패턴을 전달하는 경우
import re

# 패턴을 매번 함수에 직접 전달
text = "My phone number is 12345."

# 패턴을 매번 사용 (매번 컴파일)
for i in range(30000):
    match = re.search(r'\d+', text)
    if match:
        print("Found a number:", match.group())
```

```python
""" 
정규식 - 함수 활용
"""

# 2) 패턴을 미리 컴파일하여 객체로 만드는 경우
import re

# 컴파일된 패턴
pattern = re.compile(r'\d+')

text = "My phone number is 12345."
for i in range(30000):
    match = pattern.search(text)  # 패턴을 재사용
    if match:
        print("Found a number:", match.group())
```

```python
""" 
정규식 - 함수 활용
"""
import re

s = 'I like BTS!'
print(re.sub('BTS', 'Black Pink', s))

# 숫자만 찾아서 *으로 바꿈
# \d는 메타문자인데, \는 escape 문자로 인식되는 문제가 있음
# 따라서, r을 앞에 설정하여 \를 raw string으로 설정함
s = 'My lucky number 2 7 99'
print(re.sub('[0-9]+', '*', s))  
print(re.sub(r'\d+', '*', s)) 
```

```python
""" 
정규식 - 연습
"""

import re

s = "My lucky number 2 7 99"

def hash_by_mult_and_modulo(m):        # 매개변수로 매치 객체를 받음
    n = int(m.group())                 # 매칭된 문자열을 가져와서 정수로 변환
    return str(n * 23435 % 973)        # 숫자에 해시함수 적용 후 문자열로 변환해서 반환

print(re.sub('[0-9]+', hash_by_mult_and_modulo, s))
```
